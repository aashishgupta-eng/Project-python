```python
print("====================================")
print(" Welcome to Personal Data Collector ")
print("====================================")


name = input("Enter your name: ")
age = int(input("Enter your age: "))
height = float(input("Enter your height (in meters): "))
fav_num = int(input("Enter your favourite number: "))

birth_year = 2026 - age

height_int = int(height)


print("\n========== User Information ==========")

print("Name:", name, "| Type:", type(name), "| Memory Address:", id(name))
print("Age:", age, "| Type:", type(age), "| Memory Address:", id(age))
print("Height:", height, "| Type:", type(height), "| Memory Address:", id(height))
print("Favourite Number:", fav_num, "| Type:", type(fav_num), "| Memory Address:", id(fav_num))

print("\n========== Data Processing ==========")

print("Approximate Birth Year:", birth_year)

print("\n========== Type Casting ==========")

print("Original Height:", height)
print("Height after int() conversion:", height_int)


print("\nThank you for using the Personal Data Collector!")
print("Keep learning Python. Goodbye!")
```
