```python
# ==============================
# Student Data Organizer
# ==============================

students_list = []
students_dict = {}
subjects_set = set()

print("=" * 50)
print("     WELCOME TO STUDENT DATA ORGANIZER")
print("=" * 50)
print("This program allows you to:")


while True:

    print("\n========== MENU ==========")
    print("1. Add Student")
    print("2. Display All Students")
    print("3. Update Student")
    print("4. Delete Student")
    print("5. Display Subjects Offered")
    print("6. Exit")

    choice = input("Enter your choice: ")

    # --------------------------------
    # ADD STUDENT
    # --------------------------------
    if choice == "1":

          student_id = input("Enter Student ID: ")
          name = input("Enter Student Name: ")

          age = int(input("Enter Age: "))

          grade = input("Enter Grade: ")

          dob = input("Enter Date of Birth (DD/MM/YYYY): ")

          subjects = input("Enter Subjects (comma separated): ").split(",")


          subjects = [subject.strip() for subject in subjects]


          student_tuple = (student_id, dob)


          student = {
              "info": student_tuple,
              "name": name,
              "age": age,
              "grade": grade,
              "subjects": subjects
          }


          students_list.append(student)


          students_dict[student_id] = {
              "name": name,
              "age": age,
              "grade": grade,
              "subjects": subjects
          }


          for subject in subjects:
              subjects_set.add(subject)

          print("\nStudent Added Successfully!")

    # --------------------------------
    # DISPLAY STUDENTS
    # --------------------------------
    elif choice == "2":

          if len(students_list) == 0:
              print("No student records available.")

          else:
              print("\n========== STUDENT RECORDS ==========")

              for student in students_list:

                  sid = student["info"][0]
                  dob = student["info"][1]


                  print(f"\nStudent ID : {sid}")


                  print("Name       : {}".format(student["name"]))


                  print("Age        : %d" % student["age"])

                  print(f"Grade      : {student['grade']}")
                  print("DOB        :", dob)
                  print("Subjects   :", ", ".join(student["subjects"]))

    # --------------------------------
    # UPDATE STUDENT
    # --------------------------------
    elif choice == "3":

          sid = input("Enter Student ID to Update: ")

          if sid in students_dict:

              print("\n1. Update Age")
              print("2. Update Subjects")

              update_choice = input("Choose option: ")

              if update_choice == "1":

                  new_age = int(input("Enter New Age: "))

                  students_dict[sid]["age"] = new_age

                  for student in students_list:
                      if student["info"][0] == sid:
                          student["age"] = new_age

                  print("Age Updated Successfully!")

              elif update_choice == "2":

                  new_subjects = input("Enter New Subjects (comma separated): ").split(",")

                  new_subjects = [sub.strip() for sub in new_subjects]

                  students_dict[sid]["subjects"] = new_subjects

                  for student in students_list:
                      if student["info"][0] == sid:
                          student["subjects"] = new_subjects


                  subjects_set.clear()

                  for student in students_list:
                      for subject in student["subjects"]:
                          subjects_set.add(subject)

                  print("Subjects Updated Successfully!")

              else:
                  print("Invalid Option!")

          else:
              print("Student ID Not Found!")

    # --------------------------------
    # DELETE STUDENT
    # --------------------------------
    elif choice == "4":

          sid = input("Enter Student ID to Delete: ")

          found = False

          for i in range(len(students_list)):
              if students_list[i]["info"][0] == sid:


                  del students_list[i]

                  found = True
                  break

          if found:

              if sid in students_dict:
                  del students_dict[sid]


              subjects_set.clear()

              for student in students_list:
                  for subject in student["subjects"]:
                      subjects_set.add(subject)

              print("Student Deleted Successfully!")

          else:
              print("Student ID Not Found!")

    # --------------------------------
    # DISPLAY SUBJECTS
    # --------------------------------
    elif choice == "5":

          print("\nUnique Subjects Offered")

          if len(subjects_set) == 0:
              print("No Subjects Available.")

          else:
              for subject in subjects_set:
                  print("-", subject)

    # --------------------------------
    # EXIT
    # --------------------------------
    elif choice == "6":

          print("\nThank you for using Student Data Organizer.")
          print("Have a Great Day!")
          break

    else:
          print("Invalid Choice! Please Try Again.")
```
